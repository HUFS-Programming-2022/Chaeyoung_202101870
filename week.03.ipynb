## 중첩조건문 
- if문의 중첩(if문 안에 if문)을 의미하며, 중첩조건문은 '들여쓰기'로 조건식을 구분한다.

##### 논리연산자 
- 두 개의 조건식(비교연산자가 여러 번 이용될 때 사용) 등을 결합하여 하나의 결과값을 만들 때 사용되는 연산자
- 종류:
A and B -> 두 개의 조건식이 모두 참일 때(A와 B 모두 true)  
A or B -> 두 개의 조건식 중 한 개의 식만 참이면 true (A 혹은 B 중 하나가 true)
not (A) -> A가 false면 true. true면 false
- 비교연산자의 괄호는 가독성을 위한 것이므로 선택적인 부분

<중첩 조건문을 만드는 방법 2가지>
방법01)
논리연산자를 사용하여 여러가지 조건을 한줄에 놓음
ex) if (a > 0) and (a < 8):
        print("A")
    else:
        print("B")
방법02)
파이썬에서만 가능한 방법으로, 조건을 하나로 뭉쳐 표현함
ex) if 0 < a < 8:
        print("A")
    else:
        print("B")

```python
#<연습코드>
a = 1
b = 2
if a == b:
    if a > b:
        print("A")
    else:
        print("B")
else:
    print("C")
```

```python
a = 5
if (a > 3) and (a < 7):
    print("yes")
else:
    print("no")
```

```python
b = 3
if 0 < b < 1:
    print("yes")
else:
    print("no")
```

## in 
- 어떤 요소가 그 안에 있는지를 확인하고 싶을 때 사용하는 문법 
- 비슷한 조건이 여러 번 반복될 때 역시 사용이 가능함

```python
#<연습코드>

letter = 'a'
if letter in ['a','e','i','o','u']:
    print("yes")
else:
    print("no")
```

```python
word = 'abcde'
if 'a' or 'b' or 'c' or 'd' or 'e' in word:
    print("yes")
else: 
    print("no")
```

# 바다코끼리 연산자
- 할당과 테스트를 한줄로 동시에 실행 가능한 문법
- :=(바다코끼리 연산자 표현)

```python
tweet_limit = 300
tweet_string = 'blah' * 100
tweet_string
```

```python
tweet_limit = 300
tweet_string = 'blah' * 100
diff = tweet_limit - len(tweet_string)
if diff > 0:
    print("가능")
else: print("불가능")
```

```python
tweet_limit = 300
tweet_string = 'blah' * 100
if diff := tweet_limit - len(tweet_string) > 0:
    print("가능")
else: print("불가능")
```

### Quiz 
- 윤년을 구하는 코드를 짜보자
  - 연도가 4로 나눠떨어지면 윤년
  - 100으로 나눠떨어지면 안된다 
  - 400으로 나눠떨어지면 된다,
---------------------------------------------------
 - year를 입력받는다. 
 - 출력={}년은 윤년입니다'/'{}년은 윤년이 아닙니다.

```python
#내가 쓴 코드 [f'{}~'로 숫자받기']
year = int(input("연도를 입력하세요:"))

if year % 4 == 0:
    print(f'{year}년은 윤년입니다')
elif year % 100 == 0:
    print(f'{year}년은 윤년이 아닙니다')
elif year % 400 == 0:
    print(f'{year}년은 윤년입니다')
else:
    print(f'{year}년은 윤년이 아닙니다')
```
