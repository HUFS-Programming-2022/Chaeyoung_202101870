#### list comprehension
-포함형,지능형,축약형 리스트
-하나의 리스트에 그 안에 있는 값을 포함하여 다른 리스트를 만드는 방식
- list + for loop
- [A for B in C] 대괄호 반드시 적기!!
    - C 에 있는 B에서 A를 가져와라 => 리스트에 담아서 반환
    - A:내가 원하는 출력 형태 
    - B:C의 형태
    - C:반복가능한 자료형(리스트)
- 조건 추가: [A for B in C if .....]

```python
[fruit[1] for fruit in fruits]
```

```python
[fruit for fruit in fruits]
```

###### fruits['lemon']->잘못된 표현
###### -> fruits.index('lemon')

#### cased, uncased
- 변환: upper(), lower(), title(), swapcase()
->upper():대문자로 변환하는 방식
->lower():소문자로 변환하는 방식
->title():앞쪽만 대문자로 변환하는 방식
->swapcase():대문자는 소문자로, 소문자는 대문자로 변환하는 방식
-형태: isupper(), islower(), istitle()
    - endswith(<string>), startswith()
    - isalpha(): 문자..? 
- 'string'.upper()
- 'string'.endswith()

```python
# text2에서 'ness'로 끝나는 단어를 중복없이, 정렬해서 
# list->list
text1.tokens

[word for word in text2.tokens if word.endswith('ness')]
```

```python
# text1에서 단어에 'ant'가 포함된 단어들을 소문자로 리스트 반환
text1.tokens

[word.lower() for word in text1 if 'ant'in word]
```

```python
[word.lower() for word in set(text1) if 'ant'in word]

set([word.lower() for word in text1 if 'ant'in word])
```
